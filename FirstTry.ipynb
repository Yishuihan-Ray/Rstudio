```python
%%R
rainfall.data = read.table('http://stats191.stanford.edu/data/rainfall.csv', header=TRUE, sep=',')
groundhog.data = read.table('http://stats191.stanford.edu/data/groundhog.table', header=TRUE, sep=',')
shadow = groundhog.data$shadow
rainfall = subset(rainfall.data, WY %in% 1990:2010)$Total
boxplot(rainfall ~ shadow)
```
